```python
# calcular area de un triangulo
# variables b,h,A
# Formulas A = b * h / 2
# calcular el valor de b
# calcular el valor de h
print("Calcular el area de un triangulo rectangulo")
b = float(input("Ingrese el valor de la base: "))
h = float(input("Ingrese el valor de la altura: "))
A = b * h / 2
print("El area del triangulo es: ", A)
print("Karol Castillo NL = 9")
```
